```python
df1 = pd.read_csv('results.csv')

df1
```

```python
df1.drop('neutral',axis=1,inplace=True)
```

```python
df1.head(50)
```

```python
df1.rename(columns = {"date": "Date", "home_team":"Home Team", "away_team":"Away Team", "home_score":"Home Team Score", 
                    "away_score":"Away Team Score", "tournament":"Tournament", "city":"City (Match Played)", 
                      "country":"Country (Match Played)"}, inplace = True)      
```

```python
df1
```

```python
df1['Date'] = np.where(df1['Date'].str.contains('/'), pd.to_datetime(df1['Date']).dt.strftime('%m/%d/%Y'), pd.to_datetime(df1['Date'], dayfirst=True).dt.strftime('%m/%d/%Y'))
```

```python
df1
```

```python
conditions = [
    df1['Home Team Score'] > df1['Away Team Score'],
    df1['Home Team Score'] < df1['Away Team Score']]

choices = [df1['Home Team'], df1['Away Team']]

df1['Winner'] = np.select(conditions, choices, default='Draw')
```

```python
for i in df1.index:
    for j in df2.index:
        if df2['Date'][j]==df1['Date'][i] and df2['Home Team'][j]==df1['Home Team'][i] and df2['Away Team'][j]==df1['Away Team'][i]:
            df1['Winner'][i] = df2['Winner'][j]
            df1['Outcome Decided'][i] = df2['Outcome Decided'][j]
```
